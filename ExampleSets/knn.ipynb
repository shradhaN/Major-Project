```python
X = np.array(df.drop(['7'],1))
y = np.array(df['7'])
y = y.astype("int")

X_train, X_test, y_train, y_test =train_test_split(X,y,test_size = 0.33, random_state=17)

clf = neighbors.KNeighborsClassifier()
clf.fit(X_train,y_train)

accuracy = clf.score(X_test, y_test)
print(accuracy)


```
